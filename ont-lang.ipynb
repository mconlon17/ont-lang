```python
from rdflib import Graph
g = Graph()
g = g.parse("../language-ontology/lang.owl")
```

```python
print(len(g))
```

```python
qres = g.query(
    """
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    SELECT ?s (MAX(?slang1) AS ?lang1) (MAX(?slang2) AS ?lang2)
    WHERE {
      {?s a owl:Class} UNION {?s a owl:ObjectProperty} UNION {?s a owl:DatatypeProperty} UNION {?s a owl:AnnotationProperty}
      OPTIONAL {?s rdfs:label ?label . BIND(lang(?label) AS ?lang)
        BIND(IF(?lang="en", ?label, "") AS ?slang1)
        BIND(IF(?lang="de", ?label, "") AS ?slang2)
      }
      }
    GROUP BY ?s
    ORDER BY ?s
    """

)
for row in qres:
    print("%s\t%s\t%s" % row)

with open('lang-en-de.csv', 'a') as out:
    out.write("Entity\tLabel1\tLabel2\n")
    for row in qres:
        out.write("%s\t%s\t%s\n" % row)
```
